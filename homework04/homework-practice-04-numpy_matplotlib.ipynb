#### Задание 4

```python
A = np.random.rand(5, 5)
```

```python
B = np.dot(A.T, A)
C = np.dot(A, A.T)
```

Размер матрицы B будет $5 \times 5$ и размер матрицы C также будет $5 \times 5$, так как в обоих случаях произведение матрицы на транспонированную её версию даст квадратную матрицу того же размера, что и исходная матрица.

```python
print("Размер матрицы B:", B.shape)
print("Размер матрицы C:", C.shape)
```

Матрица B получается из транспонированной матрицы A умножением на саму матрицу A. Матрица C получается умножением матрицы A на транспонированную её версию. Они различаются в порядке умножения.

```python
sum_diagonal_B = np.trace(B)
sum_diagonal_C = np.trace(C)
print("Сумма диагональных элементов матрицы B:", sum_diagonal_B)
print("Сумма диагональных элементов матрицы C:", sum_diagonal_C)
```

Сумма диагональных элементов матрицы B является скалярным произведением строк и столбцов матрицы A, т.е. это скалярное произведение вектора строк на себя, что может быть интерпретировано как сумма квадратов элементов вектора.

Сумма диагональных элементов матрицы C представляет собой скалярное произведение строк и столбцов транспонированной матрицы A, что также является суммой квадратов элементов вектора.

```python
print("Детерминант матрицы A:", np.linalg.det(A))
print("Ранг матрицы A:", np.linalg.matrix_rank(A))
```

Если ранг матрицы равен числу её строк (или столбцов), то строки (или столбцы) линейно независимы.

```python
print("Детерминант матрицы B:", np.linalg.det(B))
print("Ранг матрицы B:", np.linalg.matrix_rank(B))
```

```python
print("Детерминант матрицы C:", np.linalg.det(C))
print("Ранг матрицы C:", np.linalg.matrix_rank(C))
```

```python
eigenvalues_B = np.linalg.eigvals(B)
eigenvalues_C = np.linalg.eigvals(C)
print("Спектр матрицы B:", eigenvalues_B)
print("Спектр матрицы C:", eigenvalues_C)
```

```python
import numpy as np

student_id = 1132226463

np.random.seed(student_id)
integral_choice = np.random.randint(10)
print("Выбранный интеграл:", integral_choice)
```

```python
def f(x):
    return 1 / (x * (1 + np.log(x)**2))
```

```python
# Метод прямоугольников
def rectangle_rule(a, b, N):
    dx = (b - a) / N
    x_left = np.linspace(a, b - dx, N)
    x_right = np.linspace(a + dx, b, N)
    integral = np.sum(f((x_left + x_right) / 2) * dx)
    return integral
```

```python
# Метод трапеций
def trapezoidal_rule(a, b, N):
    dx = (b - a) / N
    x_values = np.linspace(a, b, N+1)
    integral = 0.5 * (f(a) + f(b))
    integral += np.sum(f(x_values[1:-1]))
    integral *= dx
    return integral
```

```python
# Метод Симпсона
def simpsons_rule(a, b, N):
    dx = (b - a) / N
    x_values = np.linspace(a, b, N+1)
    integral = f(a) + f(b)
    integral += 4 * np.sum(f(x_values[1:-1:2]))
    integral += 2 * np.sum(f(x_values[2:-2:2]))
    integral *= dx / 3
    return integral
```

```python
def calculate_integral(true_integral, epsilon=1e-5):
    a = 1
    b = np.e

    N = 1

    # Метод прямоугольников
    integral_rectangle = rectangle_rule(a, b, N)
    while True:
        N *= 10

        integral_new = rectangle_rule(a, b, N)
        if np.abs(integral_new - integral_rectangle) < epsilon:
            break
        integral_rectangle = integral_new

    # Метод трапеций
    N = 10
    integral_trapezoidal = trapezoidal_rule(a, b, N)
    while True:
        N *= 10
        integral_new = trapezoidal_rule(a, b, N)
        if np.abs(integral_new - integral_trapezoidal) < epsilon:
            break
        integral_trapezoidal = integral_new

    # Метод Симпсона
    N = 10
    integral_simpsons = simpsons_rule(a, b, N)
    while True:
        N *= 10
        integral_new = simpsons_rule(a, b, N)
        if np.abs(integral_new - integral_simpsons) < epsilon:
            break
        integral_simpsons = integral_new

    print("Приближенное значение интеграла методом прямоугольников:", integral_rectangle)
    print("Приближенное значение интеграла методом трапеций:", integral_trapezoidal)
    print("Приближенное значение интеграла методом правила Симпсона:", integral_simpsons)
    print("Аналитическое значение интеграла:", true_integral)
```

```python
# Аналитическое значение интеграла если посчитать своими силами
true_integral = 1.338261212718

calculate_integral(true_integral)
```
